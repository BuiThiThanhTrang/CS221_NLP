# Import model

```python
import torch
import torch.nn as nn
import torch.optim as optim
import importlib.util
import pickle
import sys
import os

# 1. Thêm đường dẫn thư mục repo vào sys.path
file_path = "/content/drive/MyDrive/CS221_NLP/CS221_NLP/BT2NLPQ-Starting-Kit/model.py"

spec = importlib.util.spec_from_file_location("model", file_path)
model_module = importlib.util.module_from_spec(spec)
sys.modules["model"] = model_module
spec.loader.exec_module(model_module)

# Bây giờ lấy class ra dùng
FirstLM = model_module.FirstLM
SecondLM = model_module.SecondLM

print("Đã load trực tiếp FirstLM và SecondLM từ file chỉ định.")
```

# Huấn luyện

```python
# 1. Khởi tạo mô hình
first = FirstLM()
second = SecondLM()

print("Đang đọc dữ liệu...")

# 2. Đọc dữ liệu từ file 1.txt
txt_1 = "/content/drive/MyDrive/CS221_NLP/CS221_NLP/BT2NLPQ-Starting-Kit/1.txt"
txt_2 = "/content/drive/MyDrive/CS221_NLP/CS221_NLP/BT2NLPQ-Starting-Kit/2.txt"
fdata = []

try:
    with open(txt_1, "rt", encoding="utf-8") as r:
        for s in r:
            fdata.append(s.strip())
except FileNotFoundError:
    print("LỖI: Không tìm thấy file 1.txt")

# 3. Đọc dữ liệu từ file 2.txt
sdata = []
try:
    with open(txt_2, "rt", encoding="utf-8") as r:
        for s in r:
            sdata.append(s.strip())
except FileNotFoundError:
    print("LỖI: Không tìm thấy file 2.txt")

# 4. Huấn luyện (Fit)
print(f"Đang huấn luyện FirstLM với {len(fdata)} dòng...")
first.fit(fdata)

print(f"Đang huấn luyện SecondLM với {len(sdata)} dòng...")
second.fit(sdata)

# 5. Lưu mô hình (Save)
print("Đang lưu mô hình ra file .mdl...")
# Gọi hàm save() đã được viết trong class (sử dụng pickle)
first.save()   # Sẽ tạo ra FirstLM.mdl
second.save()  # Sẽ tạo ra SecondLM.mdl

print("HOÀN TẤT: Đã tạo file FirstLM.mdl và SecondLM.mdl")
```
